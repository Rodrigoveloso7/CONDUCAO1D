```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,0.1,5)
T = np.zeros(5)
```

RESOLUÇÃO NUMÉRICA POR DIFERENÇAS FINITAS

image.png

```python
dx = x[1]-x[0]
q =1e5
k=400
```

```python
T_dif = [293.15, 273.15,273.15,273.15,303.15]
```

```python
T_dif
```

```python
for n in range(100):
  for i in range(1,4):
    T_dif[i] = ((-q/(k*x[i]))*(dx**2)-T_dif[i+1] -T_dif[i-1])/(-2)
```

```python
T_dif
```

```python
plt.plot(x,T_dif,color='black')
plt.title('Temperatura em função da posição - Diferenças Finitas')
plt.xlabel('x[m]')
plt.ylabel('T[K]')
plt.grid()
plt.show()
```
