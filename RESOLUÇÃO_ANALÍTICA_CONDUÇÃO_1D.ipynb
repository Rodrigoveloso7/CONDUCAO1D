```python
x = np.linspace(0,0.1,5)
T = np.zeros(5)
```

```python
x
```

```python
for i in range(1,4):
  T[i] = (-1000/4)*(np.log(x[i])*x[i]-x[i]+2.904*x[i]-1.1726)
  print(T[i])
T[0] = 293.15
T[-1] = 303.15
```

```python
plt.plot(x,T,color='black')
plt.title('Temperatura em função da posição')
plt.xlabel('x[m]')
plt.ylabel('T[K]')
plt.grid()
```
